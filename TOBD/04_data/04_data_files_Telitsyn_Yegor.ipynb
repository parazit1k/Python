### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
with open('steps_sample.xml', 'r') as f: data = BeautifulSoup(f.read(), 'xml')

result = {
    
}

for i in data.find_all('recipe'):
    result[int(i.find('id').get_text())] = []
    
    for j in i.find_all('step'):
        result[int(i.find('id').get_text())].append(j.get_text())

result
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
result = {
    
}

for i in data.find_all('recipe'):
    if len(i.find_all('step')) not in result:
        result[len(i.find_all('step'))] = []
    result[len(i.find_all('step'))].append(int(i.find('id').get_text()))

result
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
result = []

for i in data.find_all('recipe'):
    for j in i.find_all('step'):
        if j.has_attr('has_hours') | j.has_attr('has_minutes'):
            result.append(int(i.find('id').get_text()))
            
print(len(result))
print(result)
````
